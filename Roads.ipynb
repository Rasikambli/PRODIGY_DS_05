```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("C:\\Users\\RASIKA KAMBLI\\Downloads\\archive (14)\\US_Accidents_March23.csv" , nrows=1000)
```

```python
data =data.drop(columns=['ID','Source','Start_Lat','Start_Lng','End_Lat','End_Lng','Precipitation(in)' ])
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Wind_Chill(F)'].fillna(data['Wind_Chill(F)'].mean(),inplace=True)
data['Wind_Speed(mph)'].fillna(data['Wind_Speed(mph)'].mean(),inplace=True)
data['Visibility(mi)'].fillna(data['Visibility(mi)'].mean(),inplace=True)
data['Pressure(in)'].fillna(data['Pressure(in)'].mean(),inplace=True)
data['Humidity(%)'].fillna(data['Humidity(%)'].mean(),inplace=True)
data['Temperature(F)'].fillna(data['Temperature(F)'].mean(),inplace=True)
```

```python
data['Weather_Condition'] = data['Weather_Condition'].fillna(data['Weather_Condition'].mode().iloc[0])
data['Wind_Direction'] = data['Wind_Direction'].fillna(data['Wind_Direction'].mode().iloc[0])
data['Weather_Timestamp'] = data['Weather_Timestamp'].fillna(data['Weather_Timestamp'].mode().iloc[0])
```
